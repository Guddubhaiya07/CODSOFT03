```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['species'] = le.fit_transform(data['species'])
data.head()
```

```python
species_name = le.classes_
print(species_name)
```

<h3 style="color:blue">Drop to the value of Data</h3>

```python
X = data.drop(columns=['sepal_length', 'species'])
y = data['species']
X.head(3)
```

<h3 style="color:blue"> Import to (train_test_split,,,StandarScaler) in X,Y Axis</h3>

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=69)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
print(x_train[:1])
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
print(x_train[:1])
```

```python
pip install numpy seaborn tensorflow
```

```python
import numpy as np
import seaborn as sns
from tensorflow import keras
```

```python
y_train = keras.utils.to_categorical(y_train, num_classes=3)
print(y_train[:5])
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_shape=(x_train.shape[-1], )))
model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=3, activation='softmax'))
```

<h3 style="color:blue">Summary</h3>

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train, y_train, epochs=100, verbose=2)
```

<h3 style="color:blue">prediction of the data in x_test</h3>

```python
prediction = model.predict(x_test)
print(prediction[:5])
```

```python
prediction = np.argmax(prediction, axis=-1)
print(prediction[:5])
```

```python
print(y_test[:5])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(accuracy_score(y_test, prediction))
```

```python
cm = confusion_matrix(y_test, prediction)
print(cm)
```

<h3 style="color:blue">confusion matrix for irsis dataset prediction</h3>

```python
ax = sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', xticklabels=species_name, yticklabels=species_name)
ax.set_title('confusion matrix for irsis dataset prediction')
ax.set_xlabel('prediction', fontsize=14)
ax.set_ylabel('actual', fontsize=14)
```
